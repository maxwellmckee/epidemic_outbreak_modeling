# Epidemic Outbreak Modeling

### Part 1: Outbreak Modeling
This part involves modeling an outbreak within a school community, represented by students, teachers, and staff. This portion includes constructing an undirected graph to simulate an outbreak under the SIS model, starting from a specified "patient-0". The simulation results, including the number of infected and susceptible individuals over time, are visualized to understand the outbreak's dynamics. Additionally, this fits an exponential growth model to the early phase of the outbreak, comparing theoretical and empirical growth rates, and analyzing the community's structure for varying transmission and recovery rates.

### Part 2: Transmission Rate Analysis
The focus shifts to examining how varying the transmission rate affects the spread of infection while keeping the recovery rate constant. Through a series of simulations across a range of transmission rates, this part evaluates the average time constant $\tau$ and endemic size, providing insights into the conditions necessary for an epidemic to occur. Theoretical and experimental analyses are compared to understand the impact of transmission rates on the outbreak's severity and persistence.

### Part 3: Influence of Patient-0 Centrality on Outbreak Spread
Exploring the impact of the initial infection point, this section investigates how the centrality of "patient-0" affects the speed and extent of the outbreak. By simulating the outbreak from every node in the network and analyzing various centrality measures (degree, closeness, betweenness, eigenvector), this part seeks to identify which metrics are better predictors of an outbreak's spread. The relationship between centrality and the outbreak's dynamics is quantified through Pearson correlation coefficients, offering a nuanced understanding of how network structure influences disease transmission.

```python
import time

import EoN as eon
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import scipy
from tqdm.notebook import tqdm

from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
```

## Part 1: Outbreak Modeling

```python
def load_flu_network():
    # Read the graph
    G = nx.read_edgelist("fludata.txt", nodetype=int, data=(("weight", float),))

    return G
```

```python
def simulate_outbreak(G, n_iter, initial_infected=325, tmax=10, beta=0.01, mu=0.5):
    """
    Inputs:
        G : nx.Graph
        n_iter : int
              Number of simulations to run.
        initial_infected : int
                        The node that is initially infected
        tmax : int
        beta : float
        mu : float

    Returns:
        simulation_runs : list[tuple] -
                       This is a list like object of tuples of the
                       form (t, S, I). Each tuple (t, S, I) represents
                       a simulated run, where t, S, I are vectors representing
                       the time periods, and the corresponding numbers of susceptible
                       and infected people.
    """
    simulation_runs = []

    for simulation in range(n_iter):
        # Fast_SIS simulation
        result = eon.fast_SIS(G, tau=beta, gamma=mu, initial_infecteds=initial_infected, 
                              tmax=tmax, return_full_data=False)
        times, S, I = result # Result is tuple

        if I[-1] != 0:  # Check last # of infected
            simulation_runs.append(result)

    # Filter out sims where outbreak died
    simulation_runs = [run for run in simulation_runs if run[2][-1] != 0]

    return simulation_runs


def plot_outbreaks(simulation_runs, save=False):
    """
    Inputs:
        simulation_runs : list[tuple]
                      The list like object of tuples returned by `simulate_outbreak`.
        save : bool
           Whether to save the figure.
    """
    plt.figure(figsize=(14, 7))

    for run_id, (t, S, I) in enumerate(simulation_runs):
        plt.plot(t, S, label=f'Run {run_id + 1} - Susceptible')
        plt.plot(t, I, linestyle='--', label=f'Run {run_id + 1} - Infected')

    plt.xlabel('Time')
    plt.ylabel('Number of People')
    plt.title('SIS Model Simulation of Disease Spread')
    plt.legend()
    
    if save:
        plt.savefig("1_1.png")

    plt.show()
```

```python
def exponential_model(t, I0, tau):
    return I0 * np.exp(t / tau)
```

```python
def get_exponent(simulation_run, I_thrsh=100):
    """
    Inputs:
        simulation_run : tuple
                      A tuple of (t, S, I) as specified in `simulate_outbreak()`
        I_thrsh : int
               Threshold of I. We only fit the curve where I<=I_thrsh

    Returns:
        tau : float
    """
    t, S, I = simulation_run  # Unpack tuple

    # Filter data to only include the exponential growth phase
    t_filtered = np.array(t)[np.array(I) <= I_thrsh]
    I_filtered = np.array(I)[np.array(I) <= I_thrsh]

    if len(t_filtered) == 0:  # Check if filtered data is empty
        return float('nan')  # Return NaN if no data to fit

    # Exponential growth model I(t) = I_0 * exp(t/tau)
    # Set initial guess such that tau is the time it takes to reach I_thrsh starting from 1 infected.
    # Estimate by assuming constant growth over observed window.
    initial_guess_tau = t_filtered[-1] / np.log(I_thrsh)
    initial_guess_I0 = 1  

    popt, pcov = curve_fit(lambda t, tau: exponential_model(t, initial_guess_I0, tau),
                           t_filtered, I_filtered, p0=(initial_guess_tau))
    tau = popt[0]  # Estimated tau

    return tau

def plot_curve_fit(simulation_run, tau, I_thrsh=100, save=False):
    """
    Inputs:
        simulation_run : tuple
        tau : float
        save : bool
    """
    t, S, I = simulation_run
    t_model = np.array(t)[np.array(I) <= I_thrsh]  # Time for modeling
    I_actual = np.array(I)[np.array(I) <= I_thrsh]  # Actual infected for modeling

    # Calculate model values using the fitted tau
    I_model = exponential_model(t_model, 1, tau)  # Theoretical predictions with fitted tau

    # Compute R-squared value
    r2 = r2_score(I_actual, I_model)

    plt.figure(figsize=(10, 6))
    plt.plot(t, I, 'o', label='Actual Infected')
    plt.plot(t_model, I_model, 'r-', label='Fitted Exponential Model')
    plt.xlabel('Time')
    plt.ylabel('Number of Infected Individuals')
    plt.title(f'Fit of exponential model to outbreak data (Tau={tau:.2f}, R^2={r2:.2f})')
    plt.legend()

    if save:
        plt.savefig("1_2.png")

    plt.show()
    
    return r2
```

```python
def calculate_theoretical_taus(G, beta=0.01, mu=0.5):
    """
    Inputs:
        G : nx.Graph
        beta : float
        mu : float

    Returns:
        tau_rand : float
        tau_slide : float
        tau_book : float
    """
    m = G.number_of_edges()
    n = G.number_of_nodes()
    k_bar = 2 * m / n  # average degree 

    # for random distribution, use k_bar(k_bar+1)
    k_bar_squared_rand = k_bar * (k_bar + 1)

    # For arbitrary distribution, calculate k_bar^2 using degree distribution
    degrees = [deg for node, deg in G.degree()]
    pk = np.bincount(degrees) / n  # probability of each degree
    k_values = np.arange(len(pk))
    k_bar_squared_arb = sum(k_values**2 * pk)  # sum of k^2 * p_k

    # calculate tau for random distribution
    tau_rand = k_bar / (beta * k_bar_squared_rand - (beta + mu) * k_bar)

    # calculate tau for arbitrary distribution using slide definition (same as random)
    tau_slide = k_bar / (beta * k_bar_squared_arb - (beta + mu) * k_bar)

    # calculate tau for arbitrary distribution using book definition
    tau_book = k_bar / (beta * k_bar_squared_arb - mu * k_bar)

    return tau_rand, tau_slide, tau_book


def compare_taus(empirical_taus, tau_rand, tau_slide, tau_book, save=False):
    """
    This function visually compares the empirical_taus, tau_rand, tau_slide,
    and tau_book on a boxplot.

    Inputs:
        empirical_taus : list[float]
                      List-like object of floats containing the list of tau's
        tau_rand : float
        tau_slide : float
        tau_book : float
        save : bool
    """
    plt.figure(figsize=(10, 6))

    # Create boxplot for empirical taus
    plt.boxplot(empirical_taus, positions=[1], widths=0.6)

    # Overlay theoretical tau values as dots
    theoretical_taus = [tau_rand, tau_slide, tau_book]
    labels = ['Random', 'Slide', 'Book']
    colors = ['red', 'green', 'blue']  # Different color for each theoretical value

    for i, tau in enumerate(theoretical_taus):
        plt.plot(1, tau, 'o', label=f'{labels[i]} (tau={tau:.2f})', color=colors[i])

    plt.xticks([1], ['Empirical Taus'])
    plt.ylabel('Tau')
    plt.title('Comparison of Empirical and Theoretical Tau Values')

    # Move legend so it doesn't block data
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
    if save:
        plt.savefig("1_3.png")

    plt.show()
```

```python
def calculate_theoretical_endemic_size(G, beta=0.01, mu=0.5):
    """
    Inputs:
        G : nx.Graph
        beta : float
        mu : float

    Returns:
        theoretical_endemic_size : int
    """
    # Calculate average degree k_hat of G.
    m = G.number_of_edges()
    n = G.number_of_nodes()
    k_hat = 2 * m / n  # average degree
    
    # Check the condition to determine if an outbreak will occur
    if k_hat * beta > mu:
        # If an outbreak occurs, the endemic size is given by the formula:
        i_endemic = 1 - (mu / (k_hat * beta))
        theoretical_endemic_size = int(round(i_endemic * n))  # convert fraction to number
    else:
        # If no outbreak occurs, the disease dies out and the endemic size is zero
        theoretical_endemic_size = 0

    return theoretical_endemic_size # Assignment says return as percent of pop, but staff on forums say integer


def compare_endemic_sizes(
    empirical_endemic_sizes, theoretical_endemic_size, save=False
):
    """
    Inputs:
        empirical_endemic_sizes : list[int]
                               List-like object of int's representing a distribution of
                               empirical endemic sizes from multiple simulation runs.
                               Refer to cell 1.5 to see how it works.
        theoretical_endemiz_size : int
        save : bool
    """
    plt.figure(figsize=(10, 6))
    
    # Boxplot for empirical data
    plt.boxplot(empirical_endemic_sizes, positions=[1], widths=0.6)

    # Overlay theoretical endemic size as a dot
    plt.plot(1, theoretical_endemic_size, 'ro', label=f'Theoretical Endemic Size')

    plt.xticks([1], ['Empirical Endemic Sizes'])
    plt.ylabel('Endemic Size')
    plt.title('Comparison of Empirical and Theoretical Endemic Sizes')
    plt.legend()
    
    if save:
        plt.savefig("1_4.png")

    plt.show()
```

```python
G = load_flu_network()

# Generate 10 simulation runs for plotting
simulation_runs = simulate_outbreak(G, 10)

plot_outbreaks(simulation_runs)

tau = get_exponent(simulation_runs[0])
r2 = plot_curve_fit(simulation_runs[0], tau)
print("> Part 1.2 <")
print(f"tau={tau:.2f}, R2={r2:.4f}")

simulation_runs = simulate_outbreak(G, 25)

empirical_taus = []
empirical_endemic_sizes = []
for run in simulation_runs:
    empirical_taus.append(get_exponent(run))
    empirical_endemic_sizes.append(run[2][-1])

tau_rand, tau_slide, tau_book = calculate_theoretical_taus(G)
print("\n> Part 1.3 <")
print(f"tau_rand={tau_rand:.4f}, tau_slide={tau_slide:.4f}, tau_book={tau_book:.4f}")

print("Empirical distribution of tau")
def describe(x):
    print(
        "Count={count}\tMean={mean:.3f}\tstd={median:.3f}\n"
        "Min={min:.3f}\t25%={p25:.3f}\t50%={p50:.3f}\t75%={p75:<.3f}\tMax={max:.3f}".format(
            count=len(x),
            mean=np.mean(x),
            median=np.std(x),
            min=min(x),
            p25=np.percentile(x, 25),
            p50=np.median(x),
            p75=np.percentile(x, 75),
            max=max(x),
        )
    )
describe(empirical_taus)

compare_taus(empirical_taus, tau_rand, tau_slide, tau_book)

print("\n> Part 1.4 <")
theoretical_endemic_size = calculate_theoretical_endemic_size(G)

compare_endemic_sizes(empirical_endemic_sizes, theoretical_endemic_size)
print(f"Theoretical endemic size={theoretical_endemic_size:.2f}")
print("Empirical distribution of simulated endemic size")
describe(empirical_endemic_sizes)
```

## Part 2: Transmission Rate Variation  $\beta$ 

### Minimum Transmission Rate for Epidemic

```python
def simulate_beta_sweep(
    G,
    n_sims,
    beta_min=0.001,
    beta_max=0.04,
    beta_samples=40,
    initial_infected=325,
    tmax=10,
    mu=0.5,
):
    """
    Generate a list of betas. Run multiple simulations for each beta.
    Save the results.

    Inputs:
        G : nx.Graph
        n_sims : int
              Number of simulations (or runs) for each beta value
        beta_min : float
                Minimum beta to simulate
        beta_max : float
                Maximum beta to simulate
        beta_samples : int
                    The number of betas to simulate. That is, the function will
                    generate `beta_samples` betas between `beta_min` and `beta_max`.
        initial_infect : int
                      Initial infected node
        tmax : int
        mu : float

    Returns:
        betas : list[float]
             The list of betas the function has generated.
        beta_runs : list[list[tuple]]
                 Simulation results corresponding to `betas`. It is a list where
                 each itemis a list of tuples, representing multiple simulations
                 for a particular beta. The tuple is in the form of (t, S, I) as
                 in `simulate_outbreak` in Part 1.
    """
    betas = np.linspace(beta_min, beta_max, beta_samples)  # Generate beta values
    beta_runs = []  # Store simulation results for each beta

    # Run simulations for each beta
    for beta in betas:
        runs_for_beta = []  # Store simulations for current beta
        for _ in range(n_sims):  # Perform n_sims simulations
            simulation_run = simulate_outbreak(G, 1, initial_infected, tmax, beta, mu)
            if simulation_run:  # Only store if the simulation didn't die out
                runs_for_beta.append(simulation_run[0])  # Unpack single simulation run result
        beta_runs.append(runs_for_beta)  # Store all runs for current beta
        
    betas = betas.tolist()

    return betas, beta_runs


def extract_average_tau(beta_runs):
    """
    Estimate the average tau value for each beta value.

    Inputs:
        beta_runs : list[list[tuple]]
                 See docstring of `simulate_beta_sweep` above.

    Returns:
        avg_taus : list[float]
                A list like object, where each item is the estimated
                average tau values of the corresponding simulations in
                `beta_runs` for a particular beta.
    """
    avg_taus = []  # Average tau's for each set of runs

    # Calculate average tau for each beta
    for runs in beta_runs:
        taus = []  # Store individual taus for runs at current beta
        for run in runs:
            tau = get_exponent(run)  # Use get_exponent to find tau for current run
            taus.append(tau)  # Store tau
        avg_taus.append(np.mean(taus))  # Store average tau for current beta

    return avg_taus


def plot_beta_tau_curves(betas, avg_taus, t, save=False):
    """
    Inputs:
        betas : list[float]
        avg_taus : list[float]
        t : list[float]
        save: bool
    """
    plt.figure(figsize=(10, 6))
    
    # Plot the exponential growth curves for each beta value using avg_taus
    for i, beta in enumerate(betas):
        # Calculate the number of infected over time based on the average tau provided
        infected_over_time = np.exp(t / avg_taus[i])  # This corrects the exponential term
        plt.plot(t, infected_over_time, label=f'Beta: {beta:.4f}, Avg. Tau: {avg_taus[i]:.2f}')

    plt.yscale('log')  # Set the y-axis to log scale, note: exponentials on log scale may look linear
    plt.xlabel('Time')
    plt.ylabel('Number of Infected Individuals (log scale)')
    plt.title('Number of Infected over Time for Varying Transmission Rates')
    plt.legend()
    
    if save:
        plt.savefig("2_1.png")

    plt.show()
```

```python
def extract_average_endemic_size(beta_runs):
    """
    Given the simulated data, compute the average size of the endemic state for
    each beta value using the simulation data in `beta_runs`.

    Inputs:
        beta_runs : list[list[tuple]]
                 See docstring of `simulate_beta_sweep()` above.

    Returns:
        avg_ends : list[float]
                A list-like object of floats. Each item is the average endemic size
                for a particular beta value estimated using the corresponding simulation
                results in `beta_runs`.
    """
    avg_endemic_sizes = []
    for runs in beta_runs:  # runs is list of sims for a specific beta
        endemic_sizes = [run[2][-1] for run in runs if run[2][-1] > 0]  # run[2] is the list of I values for the run
        if endemic_sizes:  # Avoid division by zero if all runs die out
            avg_endemic_size = np.mean(endemic_sizes)
        else:
            avg_endemic_size = 0  # If all sims die out, endemic size is zero
        avg_endemic_sizes.append(avg_endemic_size)
    return avg_endemic_sizes


def calculate_theoretical_endemic(G, betas, mu=0.5):
    """
    Compute the theoretical endemic size for each beta value in `betas`,
    assuming random distribution.

    Additionally, compute the minimum theoretical beta for
    epdemic to occur, assuming random distribution and
    arbitrary distribution respectively.

    Inputs:
        G : nx.Graph
        betas : list[float]
             List of betas used to compute `theoretical_endemics`.
             Each item of the returned `theoretical_endemics` corresponds
             to a beta value in the `betas` list at the same index.
        mu : float

    Returns:
        theoretical_endemics : list[float]
                            A list-like object of theoretical endemic sizes
                            corresponding to the betas in the `betas` parameter.
        rand_dist_min_beta : float
        arb_dist_min_beta : float
    """
    n = G.number_of_nodes()  # Total population size
    m = G.number_of_edges()  # Total number of edges
    k_hat = 2 * m / n  # Average degree
    
    # For random distribution, epidemic occurs if beta/k_hat > mu
    rand_dist_min_beta = mu / k_hat  # minimum beta for epidemic

    # For arbitrary distribution, we approximate beta when mu > beta.
    # Calculate the second moment of the degree distribution for G.
    k_square = sum(deg**2 for node, deg in G.degree()) / n
    # Approximate beta threshold for arbitrary distribution
    arb_dist_min_beta = mu * k_hat / k_square if k_square > k_hat else float('inf')  # Avoid division by zero

    # Calculate theoretical endemic sizes for given betas
    theoretical_endemic_sizes = []
    for beta in betas:
        if beta > rand_dist_min_beta:
            # Calculate endemic size using the formula I_endemic/N = 1 - mu/(beta * k_hat)
            i_endemic = (1 - mu / (beta * k_hat)) * n
            theoretical_endemic_sizes.append(max(i_endemic, 0))  # Ensure non-negative
        else:
            theoretical_endemic_sizes.append(0)  # Below threshold, no endemic size

    return theoretical_endemic_sizes, rand_dist_min_beta, arb_dist_min_beta


def compare_endemic_sizes_vs_beta(
    betas,
    avg_endemic_sizes,
    theoretical_endemic_sizes,
    rand_dist_min_beta,
    arb_dist_min_beta,
    save=False,
):
    """
    Plot a figure to visually compare the average endemic sizes from simulation vs
    the theoretical endemic sizes for different beta values. Also, it shows the minimum
    beta thresholds for an endemic to occur assuming random and arbitrary distributions
    respectively.

    Inputs:
        betas : list[float]
             A list-like object of betas corresponding to `avg_endemic_sizes` and `theoretical_endemic_sizes`
        avg_endemic_sizes : list[float]
                        The average endemic sizes as returned by
                        `extract_average_endemic_size`. Each element
                        corresponds to a beta value in the `betas` parameter
                        at the same index.
        theoretical_endemic_sizes : list[float]
                                  The theoretical endemic sizes. Each element
                                  corresponds to a beta value in the `betas`
                                  parameter at the same index.
        rand_dist_min_beta : float
        arb_dist_min_beta : float
        save : bool
    """
    plt.figure(figsize=(10, 6))
    plt.plot(betas, avg_endemic_sizes, 'o-', label='Average Endemic Sizes from Simulation')
    plt.plot(betas, theoretical_endemic_sizes, 'x-', label='Theoretical Endemic Sizes')

    # Vertical lines for minimum beta thresholds
    plt.axvline(x=rand_dist_min_beta, color='r', linestyle='--', label='Random Dist Min Beta')
    plt.axvline(x=arb_dist_min_beta, color='g', linestyle='-.', label='Arbitrary Dist Min Beta')

    plt.xlabel('Beta (Transmission Rate)')
    plt.ylabel('Endemic Size (Number of Infected Individuals)')
    plt.title('Endemic Sizes vs. Transmission Rate (Beta)')
    plt.legend()
    
    if save:
        plt.savefig("2_2.png")

    plt.show()
```

```python
G = load_flu_network()

betas, beta_runs = simulate_beta_sweep(G, 5, beta_samples=20)

avg_taus = extract_average_tau(beta_runs)
times = np.linspace(0, 2.2, 100)

plot_beta_tau_curves(betas, avg_taus, t=times)

avg_endemic_sizes = extract_average_endemic_size(beta_runs)

(
    theoretical_endemics_sizes,
    rand_dist_min_beta,
    arb_dist_min_beta,
) = calculate_theoretical_endemic(G, betas)

compare_endemic_sizes_vs_beta(
    betas,
    avg_endemic_sizes,
    theoretical_endemics_sizes,
    rand_dist_min_beta,
    arb_dist_min_beta,
)

print(">>>>> Results for Part 2 <<<<<")
with np.printoptions(precision=2, suppress=True):
    print(f"Avg_taus = {np.array(avg_taus)}\n")

with np.printoptions(precision=0, suppress=True):
    print(f"Avg endemic size = {np.array(avg_endemic_sizes)}\n")
    print(f"Theo endemic size = {np.array(theoretical_endemics_sizes)}\n")

print(f"Min beta for random distribution = {rand_dist_min_beta:.5f}")
print(f"Min beta for arbitrary distribution = {arb_dist_min_beta:.5f}")
```

### Part 2 Visualization Comments
The theoretical endemic sizes are very similar to the experimental endemic sizes. The logarithmic infection spread follows a very similar path and shape. The experimental endemic development has more variability at the start of the spread, but as you get towards the elbow of the spread they are closely aligned. Both the minimum beta values for the random distribution and arbitrary distribution are very similar. They approximate the start of the endemic well.

## Part 3

```python
def sweep_initial_infected(G, tmax=10, beta=0.01, mu=0.5):
    """
    Compute the taus for each node.

    Inputs:
        G : nx.Graph
        tmax : int
        beta : float
        mu : float
    Returns:
        taus : list[float]
            List of tau's corresponding to the `nodes` return value below.
        nodes : list[int]
    """
    taus = []
    nodes = []

    for node in G.nodes():
        # Run sim with the current node as the initial infected
        simulation_runs = simulate_outbreak(G, 1, node, tmax, beta, mu)

        # Check if we have at least one sim run that didn't die out
        if simulation_runs:  # If list is not empty
            run = simulation_runs[0]  # Get the first (and only) simulation run
            tau = get_exponent(run)  # Calculate tau using the provided function
            if tau:  # If tau calculation was successful
                taus.append(tau)  # Append tau to list
                nodes.append(node)  # Append node to list

    return taus, nodes


# Compute centrality metrics
def compute_centrality(G, nodes):
    """
    Inputs:
        G : nx.Graph
        nodes : list[int]
             A list of nodes for which we compute centralities.
    Returns:
        cent_dict : dict[list[float]]
                  The keys of the dict are 'deg', 'clo', 'bet' and 'eig'.
                  The values are lists of floats representing the corresponding
                  centrality for each node in the `nodes` parameter.

    """
    # Compute centrality measures for the given nodes
    # Degree Centrality for G
    deg_cen = nx.degree_centrality(G)
    # Closeness Centrality for G, wf_improved set to False
    clo_cen = nx.closeness_centrality(G, wf_improved=False)
    # Betweenness Centrality for G
    bet_cen = nx.betweenness_centrality(G)
    # Eigenvector Centrality for G
    eig_cen = nx.eigenvector_centrality_numpy(G)

    # Filter centralities for the nodes where sim was successful
    cent_dict = {
        "deg": [deg_cen[node] for node in nodes],
        "clo": [clo_cen[node] for node in nodes],
        "bet": [bet_cen[node] for node in nodes],
        "eig": [eig_cen[node] for node in nodes],
    }

    return cent_dict
```

```python
def calculate_person_correlation(taus, cent_dict):
    """
    Inputs:
        taus : list[float]
        cent_dict : dict[list[float]]
                 See docstring in `compute_centrality`

    Returns:
        r_dict : dict[tuple[float, float]]
              The keys of the dict are 'deg', 'clo', 'bet' and 'eig'.
              The values are tuples of (Pearson coefficient, p-value).
    """
    r_dict = {}
    # Calculate Pearson correlation coefficient and p-value for each centrality metric
    for cent_type in cent_dict.keys():  # 'deg', 'clo', 'bet', 'eig'
        r_value, p_value = pearsonr(cent_dict[cent_type], taus)
        r_dict[cent_type] = (r_value, p_value)
    return r_dict


def plot_centrality_vs_tau(taus, cent_dict, r_dict, save=False):
    """
    Inputs:
        taus : list[float]
            See docstring in `sweep_initial_infected`
        cent_dict : dict[list[float]]
                 See docstring in `compute_centrality`
        r_dict : dict[tuple[float, float]]
              See docstring in  `calculate_person_correlation`

        save: bool
    """
    # Plot centrality metrics against tau (4 plots)
    fig, axs = plt.subplots(2, 2, figsize=(14, 10))  # 2x2 grid
    axs = axs.flatten()  
    centrality_labels = {'deg': 'Degree', 'clo': 'Closeness', 
                         'bet': 'Betweenness', 'eig': 'Eigenvector'}

    for i, (cent_type, label) in enumerate(centrality_labels.items()):
        axs[i].scatter(cent_dict[cent_type], taus)
        axs[i].set_title(f'{label} Centrality vs Tau\n' 
                         f'Pearson r: {r_dict[cent_type][0]:.3f}, '
                         f'p-value: {r_dict[cent_type][1]:.3f}')
        axs[i].set_xlabel(f'{label} Centrality')
        axs[i].set_ylabel('Tau')

    plt.tight_layout()  

    if save:
        plt.savefig("3_2.png")

    plt.show()
```

```python
start_time = time.time()
G = load_flu_network()

taus, nodes = sweep_initial_infected(G)
cent_dict = compute_centrality(G, nodes)

r_dict = calculate_person_correlation(taus, cent_dict)
plot_centrality_vs_tau(taus, cent_dict, r_dict)

print(f"Number of included nodes = {len(nodes)}")
for k in sorted(r_dict):
    coeff, pvalue = r_dict[k]
    print(f"{k} = {coeff:.4f}, pvalue = {pvalue:.4f}")

seconds_elapsed = time.time() - start_time
print(f"Elapsed time = {seconds_elapsed/60 : .2f} minutes")
```

### Part 3 Synopsis

The centrality metrics ranked from strongest to weakest negative correlation are:

* Degree Centrality (deg)
* Eigenvector Centrality (eig)
* Closeness Centrality (clo)
* Betweenness Centrality (bet)

All of them have a negative Pearson coefficient which means that as the centrality of the node increase, the characteristic time $\tau$ decreases. More intuitively, highly central nodes correspond to faster disease spreading.

It aligns with my expectations that degree centrality would be the strongest predictor since it's intuitive to think that nodes with more connections will align with a faster disease spread. Patient-0 nodes that have more direct connections would lead to a faster disease spread.

Eigenvector centrality is a close second. It similarly considers the number of connections, but also node connection quality. So, if the disease starts at nodes that are connected to other highly connected nodes, the disease will spread faster. This also makes sense.

Closeness centrality is a close third. A negative Pearson coefficient under this centrality metric indicates that nodes that, on average, can quickly reach other nodes will excel the spread of a disease, which is not unintuitive.

The weakest centrality metric is the Betweenness centrality, and is the only metric with a Pearson coefficient significantly different from the others. Betweenness centrality examines shortest paths between nodes, so possibly an infection may not take the form of spreading to node $i$ along the shortest path making for an inefficient disease spread.
